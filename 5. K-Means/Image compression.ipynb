```python
data = np.asarray(image)
og_shape = data.shape
og = data
```

```python
data = data.reshape((data.shape[0]*data.shape[1]),4)
data.shape
```

```python
data = (data/255)
data.shape
```

```python
from kmeans import KMeans
```

```python
model = KMeans()
```

```python
model.fit(data)
```

```python
parameters = model.train(64,1)
```

```python
model.predict(data[100])
```
